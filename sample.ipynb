```python
def read_key(r, key):
    t = r.type(key)

    if t == "string":
        return r.get(key)
    if t == "hash":
        return r.hgetall(key)
    if t == "list":
        return r.lrange(key, 0, -1)
    if t == "set":
        return list(r.smembers(key))
    if t == "zset":
        return r.zrange(key, 0, -1, withscores=True)

    return None


for key in r.scan_iter(match="*", count=20):
    print(key, "=>", read_key(r, key))
```
